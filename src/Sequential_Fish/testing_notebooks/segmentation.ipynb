# Open image

```python
import cellpose
import pandas as pd
import numpy as np
from Sequential_Fish.tools import open_location
```

```python
path = "/media/SSD_floricslimani/Fish_seq/Davide/2025-04-24 - HeLa R2TP_Run12/"
location = "Location-01"
Acquisition = pd.read_feather(path + "/result_tables/Acquisition.feather")
Acquisition
```

```python
image_stack = open_location(Acquisition, location)
image_stack.shape
```

```python
dapi_stack = image_stack[...,3]
dapi_stack = np.mean(dapi_stack, axis=0) #Mean on cycles
```

# drift_correction

```python
from Sequential_Fish.tools import shift_array, safe_merge_no_duplicates
from bigfish.segmentation import remove_disjoint
```

```python
Drift = pd.read_feather(path + "/result_tables/Drift.feather")
Drift = safe_merge_no_duplicates(
    Drift,
    Acquisition,
    on="acquisition_id",
    keys=["location","cycle"]
).sort_values("cycle")
Drift
```

```python
image_stack_corrected = np.zeros_like(image_stack)
image_stack_corrected.shape
```

```python
for cycle, signal in enumerate(image_stack) :
    drift = Drift.loc[(Drift["location"] == location) & (Drift["cycle"] == cycle), ["drift_z","drift_y","drift_x"]].to_numpy().squeeze()
    print(drift)
    image_stack_corrected[cycle] = shift_array(signal, *drift)
```

```python
dapi_stack_corrected = image_stack_corrected[...,3]
dapi_stack_corrected = np.mean(dapi_stack_corrected, axis=0) #Mean on cycles
```

# Segmentation

```python
nucleus_model = cellpose.models.CellposeModel(gpu= True, model_type = "nuclei")
channels = [0,0]
```

```python
nucleus_label = nucleus_model.eval(dapi_stack, diameter= 140, channels = channels, anisotropy= 3, do_3D= True)[0].astype(np.int64)
nucleus_label_corrected = nucleus_model.eval(dapi_stack_corrected, diameter= 140, channels = channels, anisotropy= 3, do_3D= True)[0].astype(np.int64)
```

```python
nucleus_label = remove_disjoint(nucleus_label)
nucleus_label_corrected = remove_disjoint(nucleus_label_corrected)
```

### Optional save

```python
np.savez(path + "/test_segmentation", nucleus_label=nucleus_label, nucleus_label_corrected=nucleus_label_corrected)
```
