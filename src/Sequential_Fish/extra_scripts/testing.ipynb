# Spots

```python
from Sequential_Fish.tools import safe_merge_no_duplicates
```

```python
Acquisition = pd.read_feather(path + f"/result_tables/Acquisition.feather")
Gene_map = pd.read_feather(path + f"/result_tables/Gene_map.feather")
Acquisition = safe_merge_no_duplicates(Acquisition, Gene_map, keys=["cycle", "color_id","target"], on=["cycle"])
Acquisition
```

```python
Spots = pd.read_feather(path + f"/result_tables/Spots.feather")
Spots = safe_merge_no_duplicates(Spots,Acquisition, on= ["acquisition_id","color_id"], keys=["location","target"])
Spots
```

```python
Spots["target"] = Spots["target"].str.replace("Barcode","")
Spots
```

```python
Spots["coordinates"] = Spots["coordinates"].apply(tuple)
grid = Spots.groupby(["location","coordinates"]).agg({
    "target" : list,
    "is_washout" : max
})
grid = grid.loc[~grid["is_washout"]]
grid["target"] = grid["target"].apply(pd.unique)
grid["dim"] = grid["target"].apply(len)
grid, grid["dim"].unique()
```
