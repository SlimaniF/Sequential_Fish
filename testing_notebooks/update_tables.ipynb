# Add wavelength to old runs

```python
import pandas as pd
import numpy as np
```

```python
path = "//media/SSD_floricslimani/Fish_seq/Davide/2024-10-09 - SeqFISH_Run10 - HeLa-Puro_POLR2//result_tables/"
Detection = pd.read_feather(path + "/Detection.feather")
Detection
```

```python
wavelengths = [555]
color_ids = [0]

Detection['color_id'] = Detection['color_id'].astype(int)
Detection['wavelength'] = 0
for color_id, wv in zip(color_ids, wavelengths) :
    loc_idx = Detection.loc[Detection['color_id'] == color_id].index
    Detection.loc[loc_idx, ["wavelength"]] = wv

Detection.value_counts(subset=["color_id","wavelength"])
```

```python
Detection.to_feather(path + "/Detection.feather")
```
